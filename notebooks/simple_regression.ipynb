```python
import pandas as pd
import statsmodels.api as sm
from openpyxl import load_workbook

# ==========================
# Load Dataset
# ==========================
df = pd.read_excel("business_regression_data.xlsx")

# ==========================
# Function for Simple Regression
# ==========================
def run_simple_regression(x_col, y_col='monthly_sales'):
    
    X = sm.add_constant(df[x_col])
    y = df[y_col]

    model = sm.OLS(y, X).fit()

    coefficient = model.params[x_col]
    intercept = model.params['const']
    r_squared = model.rsquared
    p_value = model.pvalues[x_col]

    result = {
        "Model": f"{y_col} ~ {x_col}",
        "Independent Variable": x_col,
        "Coefficient": coefficient,
        "Intercept": intercept,
        "R_squared": r_squared,
        "P_value": p_value
    }

    print("\n")
    print("="*60)
    print(f"Regression Model: {y_col} ~ {x_col}")
    print("="*60)
    print(model.summary())

    return result


# ==========================
# Run Two Models
# ==========================
result1 = run_simple_regression("marketing_spend")
result2 = run_simple_regression("footfall")

results_df = pd.DataFrame([result1, result2])

print("\nSimple Regression Summary")
print(results_df)

# ==========================
# Update regression_workbook.xlsx
# ==========================
book = load_workbook("regression_workbook.xlsx")

with pd.ExcelWriter(
    "regression_workbook.xlsx",
    engine="openpyxl",
    mode="a",
    if_sheet_exists="replace"
) as writer:

    results_df.to_excel(
        writer,
        sheet_name="Simple_Regression_Outputs",
        index=False
    )

print("\nSimple regression results saved to regression_workbook.xlsx")

# ==========================
# Business Interpretation
# ==========================
for _, row in results_df.iterrows():

    print("\n")
    print("="*60)
    print("Business Interpretation")
    print("="*60)

    print(f"Model: {row['Model']}")
    print(f"Coefficient: {row['Coefficient']:.4f}")
    print(f"R-squared: {row['R_squared']:.4f}")
    print(f"P-value: {row['P_value']:.6f}")

    if row['P_value'] < 0.05:
        print("Variable appears statistically useful.")
    else:
        print("Variable appears statistically weak.")

    print("Higher values of this variable are associated with changes in monthly sales.")
    
```
