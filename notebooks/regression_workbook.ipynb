```python
import pandas as pd
import numpy as np

# =========================
# Load Dataset
# =========================
df = pd.read_excel("business_regression_data.xlsx")

# =========================
# Original Data Sheet
# =========================
original_data = df.copy()

# =========================
# Cleaned Data Sheet
# =========================
cleaned_data = df.copy()

# Remove duplicate rows if any
cleaned_data = cleaned_data.drop_duplicates()

# Check missing values
print("Missing Values:")
print(cleaned_data.isnull().sum())

# =========================
# Dummy Variable Sheet
# =========================
dummy_data = cleaned_data.copy()

# Create dummy variables
dummy_data = pd.get_dummies(
    dummy_data,
    columns=["region", "store_type"],
    drop_first=True,
    dtype=int
)

# =========================
# Placeholder Sheets
# =========================

simple_regression_outputs = pd.DataFrame({
    "Model": [],
    "Independent Variable": [],
    "Coefficient": [],
    "Intercept": [],
    "R_squared": [],
    "P_value": []
})

multiple_regression_output = pd.DataFrame({
    "Variable": [],
    "Coefficient": [],
    "P_value": []
})

predictions_residuals = pd.DataFrame({
    "Actual_Sales": [],
    "Predicted_Sales": [],
    "Residual": []
})

model_comparison = pd.DataFrame({
    "Model_Name": [],
    "Variables_Used": [],
    "R_squared": [],
    "Significant_Variables": [],
    "Business_Usefulness": []
})

# =========================
# Save Workbook
# =========================
with pd.ExcelWriter("regression_workbook.xlsx", engine="openpyxl") as writer:
    
    original_data.to_excel(writer,
                           sheet_name="Original_Data",
                           index=False)

    cleaned_data.to_excel(writer,
                          sheet_name="Cleaned_Data",
                          index=False)

    dummy_data.to_excel(writer,
                        sheet_name="Dummy_Variables",
                        index=False)

    simple_regression_outputs.to_excel(writer,
                                       sheet_name="Simple_Regression_Outputs",
                                       index=False)

    multiple_regression_output.to_excel(writer,
                                        sheet_name="Multiple_Regression_Output",
                                        index=False)

    predictions_residuals.to_excel(writer,
                                   sheet_name="Predictions_Residuals",
                                   index=False)

    model_comparison.to_excel(writer,
                              sheet_name="Model_Comparison",
                              index=False)

print("regression_workbook.xlsx created successfully.")
```
